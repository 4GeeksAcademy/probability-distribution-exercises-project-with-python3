## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom,uniform
import numpy as np
import matplotlib.pyplot as plt

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# For the normal Distribution 
def unfirm_pmf(x, size, prob = 0.5):
    """
    Calculates the point estimate of the uniform distribution
    """
    result = uniform.pmf(k = x, n = size, p = prob, loc = 0)

    return result
def uniform_pdf(x,size,prob=0.5):
    """
    calculates the Probability density function of the Uniform Distribution
    """
    result = uniform.pmf(x=x,scale=size,loc=0)
    return result



def uniform_f(a,b,size):
    """
    The parameters are as follow:
    a: is the low value of the distribution
    b: is the high value
    size: The number of data points


    """
    samples = np.linspace(a-1,b+1,size)
    pdf = uniform.pdf(samples,loc=a,scale=(b-a))

    # Plot the uniform distribution

    plt.figure(figsize=(8, 5))
    plt.plot(samples, pdf, label=f'Uniform({a}, {b})', color='blue')
    plt.fill_between(samples, pdf, where=((samples >= a) & (samples <= b)), color='skyblue', alpha=0.4)
    plt.title('Uniform Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
    plt.axvline(a, color='red', linestyle='--', label=f'a = {a}')
    plt.axvline(b, color='green', linestyle='--', label=f'b = {b}')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
```

```python
# For The Normal Distribution
from scipy.stats import norm


def normal_f(mean,std_value,points):
    """
    The parameters are as follow:
    mean: The mean of the distribution
    std_value: The Standard deviation of the distribution
    size: The number of data points.


    """
    samples = np.linspace(mean -4*std_value,mean +4*std_value,points) # Generates values that will fall within 4 sigma of the mean.
    pdf = norm.pdf(samples,loc=mean,scale=std_value)

    # Plot the uniform distribution

    plt.figure(figsize=(8, 5))
    plt.plot(samples, pdf, label=f'Normal({mean}, {std_value})', color='blue')
    plt.fill_between(samples, pdf, where=((samples >= mean - 3 * std_value) & (samples <= mean + 3 * std_value)), 
                     color='skyblue', alpha=0.4, label='3 Range')
    plt.title('Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean}')
    plt.axvline(mean - std_value, color='orange', linestyle='--', label=f'- = {mean - std_value}')
    plt.axvline(mean + std_value, color='orange', linestyle='--', label=f'+ = {mean + std_value}')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
```

```python



def density_normal(x, mean, std):
    """
    Calculates the density of the normal distribution
    Parameters:
    x- values of x as a numpy array
    mean- mean 
    std - Standard Deviation
    """
    result = norm.pdf(x, loc = mean, scale = std)

    return result

def cdf_normal(x, mean, std):
    """
    Calculates the cumulative density of the normal distribution
    Parameters:
    x- values of x as a numpy array
    mean- mean 
    std - Standard Deviation
    """
    result = norm.cdf(x = x, loc = mean, scale = std)

    return result



def quantile_function_normal(p, mean = 0, std = 1):
    """
    Returns the Probability Point Function of the Normal Distribution.
    This is the percentile funciton. 

    Parameters:
    p- percentile to be calculated
    mean- Mean of the distribution
    std - Standard Deviation.

    """
    result = norm.ppf(q = p, loc = mean, scale = std)

    return result

def random_normal(n, mean, std):
    """
    Generates random samples from a normal distribution
    n- number of samples
    mean- mean of the distribution
    std - Standard Deviation of the Distribution

    """
    result = norm.rvs(size = n, loc = mean, scale = std)

    return result


x= np.linspace(-10,10,1000)
density_normal(x,0,1)

fig,axis = plt.subplots(3,1,figsize=(5,7))

axis[0].plot(x,density_normal(x,0,1))
axis[0].set_xlabel("x")
axis[0].set_ylabel("Probability Density Function")
axis[0].set_title("The Normal Distribution")

axis[1].plot(x,cdf_normal(x,0,1))
axis[1].set_xlabel("x")
axis[1].set_ylabel("Cumulative Density Function")
#


axis[2].plot(x,quantile_function_normal(x,0,1))
axis[2].set_xlabel("x")
axis[2].set_ylabel("Quantile Function")
#axis[2].set_title("The Normal Distribution")


plt.show()




```
